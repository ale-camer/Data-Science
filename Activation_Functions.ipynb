```python
import numpy as np
import matplotlib.pyplot as plt
from math import e
import warnings
warnings.filterwarnings("ignore")

def sigmoid(z): return 1 / (1 + np.exp(-z))
def relu(z): return np.maximum(0, z)
def linear(z): return z
def softplus(z): return np.log(1 + e**z)
def binaryStep(z): return np.where(z<0,0,1)
def leakyrelu(z): return np.where(z<0,0.1*z,z)
def paramrelu(z): return np.where(z>=0,z,)
def comploglog(z): return 1-np.exp(-np.exp(z))
def bipolarSigmoid(z): return (1-np.exp(-z))/(1+np.exp(-z))
def lecuntanh(z): return 1.7159*np.tanh((2/3)*z)
def hardtanh(z): return np.maximum(-1, np.minimum(1,z))
def smoothrelu(z): return np.log(1 + np.exp(z))
def logit(z): return np.log(z/(1-z))
def gaussian(z): return np.exp(-1/2*(z**2))
def derivative(f, x, eps=0.000001): return (f(x + eps) - f(x - eps))/(2 * eps)
```

```python
z, nrow, ncol, xmin, xmax, ymin, ymax, alpha = np.linspace(-5, 5, 200), 5, 4, -5, 5, -4, 4, 0.5
functionNames, functions = ["Step","Sigmoid","Tanh","ReLU","Linear","SoftPlus","Binary","Leaky ReLU","Complementary log-log","Bipolar Sigmoid","LeCun's Tanh","Hard Tanh","Absolute","Smooth ReLU","Logit","Cosine","Gaussian","Sine","Inverse Tangent","Angle"], \
  [np.sign(z), sigmoid(z), np.tanh(z), relu(z), linear(z), softplus(z), binaryStep(z), leakyrelu(z), comploglog(z), bipolarSigmoid(z), lecuntanh(z), hardtanh(z), abs(z), smoothrelu(z), logit(z), np.cos(z), gaussian(z), np.sin(z), np.arctan(z), np.angle(z)]

plt.figure(figsize=(21,14))
for a, b, c in zip(range(len(functionNames)),functionNames,functions):
  plt.subplot(nrow,ncol,a+1)
  plt.plot(z, c, c='mediumblue', linewidth=1, label=b, alpha=1)
  plt.vlines(0,ymin,ymax,'black','dashed',alpha=alpha)
  plt.hlines(0,xmin,xmax,'black','dashed',alpha=alpha)
  plt.legend(loc="best", fontsize=14)
  plt.axis([xmin, xmax, ymin, ymax])
plt.suptitle("Activation functions", fontsize=25)
plt.tight_layout(rect=[0, 0, 1, 0.95])   
plt.show()
```

```python
z, nrow, ncol, xmin, xmax, ymin, ymax, alpha = np.linspace(-5, 5, 200), 5, 4, -5, 5, -2, 6, 0.5
functionNames, functions = ["Step","Sigmoid","Tanh","ReLU","Linear","SoftPlus","Binary","Leaky ReLU","Complementary log-log","Bipolar Sigmoid","LeCun's Tanh","Hard Tanh","Absolute","Smooth ReLU","Logit","Cosine","Gaussian","Sine","Inverse Tangent","Angle"], \
  [derivative(np.sign, z), derivative(sigmoid,z), derivative(np.tanh,z), derivative(relu,z), derivative(linear,z), derivative(softplus,z), derivative(binaryStep,z), derivative(leakyrelu,z),
   derivative(comploglog,z), derivative(bipolarSigmoid,z), derivative(lecuntanh,z), derivative(hardtanh,z), derivative(abs,z), derivative(smoothrelu,z), derivative(logit,z), 
   derivative(np.cos,z), derivative(gaussian,z), derivative(np.sin,z), derivative(np.arctan,z), derivative(np.angle,z)]

plt.figure(figsize=(21,14))
for a, b, c in zip(range(len(functionNames)),functionNames,functions):
  plt.subplot(nrow,ncol,a+1)
  plt.plot(z, c, c='mediumblue', linewidth=1, label=b, alpha=1)
  plt.vlines(0,ymin,ymax,'black','dashed',alpha=alpha)
  plt.hlines(0,xmin,xmax,'black','dashed',alpha=alpha)
  plt.legend(loc="best", fontsize=14)
  plt.axis([xmin, xmax, ymin, ymax])
plt.suptitle("Derivatives", fontsize=25)
plt.tight_layout(rect=[0, 0, 1, 0.95])   
plt.show()
```
